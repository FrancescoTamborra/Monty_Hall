The Monty Hall problem:
https://en.wikipedia.org/wiki/Monty_Hall_problem

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def play(strategy,tries): 
    winning_doors = np.random.randint(0, 3, tries) 
    first_choice = np.random.randint(0, 3, tries) 

    if strategy == 'KEEP':
        return first_choice == winning_doors
    elif strategy == 'CHANGE':
        return first_choice != winning_doors
```

```python
tries = 10000
print("Keeping door produces {} wins out of {} games".format(np.sum(play('KEEP',tries)),tries))
print("Changing door produces {} wins out of {} games".format(np.sum(play('CHANGE',tries)),tries))
```

## Some noice plots

```python
figure = plt.figure()
tries = 10000
plt.title('Monty Hall game results for {} games'.format(tries))
plt.xlabel("player's strategy")
plt.ylabel("number of wins %")
plot = plt.bar([1,2],[((np.mean(play('KEEP',tries)))*100), 
                      ((np.mean(play('CHANGE',tries)))*100)],tick_label=["Keep","Change"], color=["blue", "red"], hatch="//")
```

```python
wins_keeping_evolution = []
wins_changing_evolution = []
samples = [1000, 10000, 30000, 50000]
for tries in samples:
    wins_keeping_evolution.append(np.sum(play('KEEP',tries)))
    wins_changing_evolution.append(np.sum(play('CHANGE',tries)))
    
figure = plt.figure()
plt.title('Wins evolution for the two strategies')
plt.xlabel("number of games")
plt.ylabel("number of wins")
plot = plt.scatter(samples, [x for x in wins_changing_evolution], label="Change", color="red")
plot = plt.scatter(samples, [x for x in wins_keeping_evolution], label="Keep", color="blue")
legend = plt.legend()
```

```python
av_wins_keeping = []
av_wins_changing = []
for tries in samples: 
    av_wins_keeping.append(np.mean(play('KEEP',tries)))
    av_wins_changing.append(np.mean(play('CHANGE',tries)))

figure = plt.figure()
plt.title('Limit value for the two strategies')
plt.xlabel("number of games")
plt.ylabel("normalized average number of wins")
plot = plt.scatter(samples, [x for x in av_wins_changing], label="Change", color="red") 
plot = plt.scatter(samples, [x for x in av_wins_keeping], label="Keep", color="blue")
legend = plt.legend()
```
